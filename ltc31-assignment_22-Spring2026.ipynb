## Part 1: Translating DNA Sequences

a) Function: nucs_to_codons
This function takes a DNA sequence string and returns a list of codons (triplets of nucleotides). The implementation uses itertools.batched, which provides a compact and readable solution.


```python
from itertools import batched

def nucs_to_codons(dna):
    """
    Convert a DNA sequence into a list of codons.

    Parameters
    ----------
    dna : str
        A string representing a DNA sequence.

    Returns
    -------
    list of str
        A list of codon strings (length 3).
    """
    dna = dna.upper()
    return [''.join(codon) for codon in batched(dna, 3) if len(codon) == 3]
```

b) Function: to_aminoacids
This function converts a list of codons into a list of amino acids. It uses zip to explicitly represent the three nucleotide positions per codon.

```python
CODON_TABLE = {
    'TTT': 'F', 'TTC': 'F', 'TTA': 'L', 'TTG': 'L',
    'CTT': 'L', 'CTC': 'L', 'CTA': 'L', 'CTG': 'L',
    'ATT': 'I', 'ATC': 'I', 'ATA': 'I', 'ATG': 'M',
    'GTT': 'V', 'GTC': 'V', 'GTA': 'V', 'GTG': 'V',
    'TAA': '*', 'TAG': '*', 'TGA': '*'
}

def to_aminoacids(codons):
    """
    Translate a list of codons into amino acids.

    Parameters
    ----------
    codons : list of str
        A list of codon strings.

    Returns
    -------
    list of str
        A list of amino acid symbols.
    """
    amino_acids = []
    for c1, c2, c3 in zip(
        [c[0] for c in codons],
        [c[1] for c in codons],
        [c[2] for c in codons]
    ):
        codon = c1 + c2 + c3
        amino_acids.append(CODON_TABLE.get(codon, 'X'))
    return amino_acids
```

```python
c) Function: translate
This function nests the previous two functions, demonstrating how complex tasks can be broken into simpler steps.
```

```python
def translate(dna):
    """
    Translate a DNA sequence into an amino acid sequence.

    Parameters
    ----------
    dna : str
        A DNA sequence.

    Returns
    -------
    list of str
        The translated amino acid sequence.
    """
    codons = nucs_to_codons(dna)
    return to_aminoacids(codons)
```

## Part 2: Reverse Complement

```python
Function: reverse_complement
This function returns the reverse complement of a DNA sequence.
```

```python
def reverse_complement(seq):
    """
    Return the reverse complement of a DNA sequence.

    Parameters
    ----------
    seq : str
        A string representing a DNA sequence.

    Returns
    -------
    str
        The reverse complement DNA sequence.
    """
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    seq = seq.upper()
    return ''.join(complement[base] for base in reversed(seq))
```

One concept I found especially valuable is the idea of composability — building small, well-defined functions and then combining them to perform more complex tasks. This approach makes debugging easier, improves readability, and encourages code reuse. In biological data analysis, where pipelines often grow incrementally, this style of programming is especially powerful.
