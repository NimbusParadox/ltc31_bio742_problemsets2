3. This problem and the next two build on the Boolean network simulation model we examined previously.

```python

import itertools

starting_conditions = list(itertools.product([True, False], repeat=3))
starting_conditions
```

```python

def simulate(V1, V2, V3, steps=50):
    history = []

    for _ in range(steps):
        history.append((V1, V2, V3))

        V1_new = not V2
        V2_new = V1 and V3
        V3_new = V1 or V2

        V1, V2, V3 = V1_new, V2_new, V3_new

    return history
```

```python

results = []

for start in starting_conditions:
    history = simulate(*start)
    last_10 = history[-10:]

    fixed = all(state == last_10[0] for state in last_10)

    results.append([list(start), fixed])

results
```
